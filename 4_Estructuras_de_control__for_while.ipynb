# **Estructuras de control For y While**

**Iterables e iteradores**

Para entender los ciclos for, es muy importante entender los conceptos de iterables e iteradores.

Empecemos con un par de definiciones:

*     Los iterables son aquellos objetos que como su nombre indica pueden ser iterados, lo que dicho de otra forma es, que puedan ser indexados. Si piensas en un array (o una list en Python), podemos indexarlo con lista[1] por ejemplo, por lo que sería un iterable.

*     Los iteradores son objetos que hacen referencia a un elemento, y que tienen un método next que permite hacer referencia al siguiente. El iterador es el que "trabaja" para entregar los valores y es "invisible".

```python
# Contar cuantas veces se repite una letra en una cadena

cadena = 'COLEGIO DE POSTGRADUADOS'

no_veces = 0
vocal = 'O'

for letra in cadena:
    if letra == vocal:
        no_veces += 1

print(f'La letra " {vocal} " se repite {no_veces} veces')
```

# El ciclo While

La sentencia while permite repetir la ejecución de un bloque de código siempre y cuando la condición del bucle sea verdadera.

image.png

```python
# Este ciclo while imprime los números del 0 al 4
contador = 0 # Inicializamos el contador en 0
while contador < 5: # El ciclo se ejecuta mientras el contador sea menor que 5
    print(contador) # Imprimimos el valor actual del contador
    contador += 1 # Incrementamos el contador en 1

# Resultado:
# 0
# 1
# 2
# 3
# 4

# ¡Cuidado con los bucles infinitos! Si la condición nunca se vuelve falsa, el ciclo se ejecutará indefinidamente.
```

***En el ciclo while debemos asegurarnos que el bucle va a terminar en algun momento***

Como nos aseguramos que el ciclo termine?

Con el contador.


```python
# i = i + i  es igual a i += 1 que es igual a i++

i = 1 # declaramos inicio del contador
limite = 5  # declaramos el limite del bucle

while i < limite:  # condición: Mientras que i sea menor o igual a 5
    print('¡Hola mundo!', 'La i vale: ', i)
    i += 1
```

```python
# bucle infinito
# no ejecutar
i = 1 # declaramos inicio del contador
limite = 5  # declaramos el limite del bucle

while i != limite:  # condición: Mientras que i sea menor o igual a 5
#while i < limite:  # condición: Mientras que i sea menor o igual a 5
    print('¡Hola mundo!', 'La i vale: ', i)
    i += 2  # incrementa en do
```

```python
# contar hacia atras
n = 5
while n > 0:
    print(n)
    n = n-1 # n = n - 1
```

```python
# S T O P
# Con while true, aseguramos que al menos una vez, se va a ejecutar el codigo dentro del ciclo.
while True:
    linea = input('Ingresa una linea: ')
    if linea == 'fin':
        break
    print(linea)
print('Fin')
```

```python
palabra = "colegio"
posicion = 0  # Inicializamos la posición en 0
letra = "e" #definimos que letra buscamos
while posicion < len(palabra):  # Mientras la posición sea menor que la longitud de la palabra
    if palabra[posicion] == letra:  # Si el carácter en la posición actual es "e"
        print(f"La letra 'e' se encuentra en la posición: {posicion + 1}")  # Imprimir la posición (sumamos 1 porque en Python los índices comienzan en 0)
        break  # Salir del ciclo
    posicion += 1  # Incrementar la posición
else:
    print("La letra 'e' no se encuentra en la palabra.")  # Si no se encuentra "e"
```

```python
#  suma o resta con while
# Creacin d eun menú de opciones
a = 6
b = 8

opcion = 0
while True:
    print("""
    Dime, ¿Que quieres hacer?
    1) Sumar los dos numeros
    2) Restar los dos numeros
    3) Salir
    """)
    opcion = int(input())
    if opcion == 1:
        print("La suma de ", a, " + ", b, " es: ", a+b)
    elif opcion == 2:
        print("La resta de ", a, " - ", b, " es: ", a-b)
    elif opcion == 3:
        break
    else:
        print("Opcion incorrecta")
```
